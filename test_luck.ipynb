##### Success and Luck: Good Fortune and the Myth of Meritocracy

##### 실험
* 0과 98 사이의 난수를 생성하여, 재능/노력에 부여.
* 0과 2 사이의 난수를 생성하여, 행운에 부여.
* 가장 노력한 10명 모두 승자가 되는지 확인해 볼 것.

```python
# import modules
import random
import matplotlib.pyplot as plt
```

```python
def generate_top_scores(n):
    """Generate top scores based on effort and luck for n samples."""

    # Initialize scores and lists
    effort_top10, luck_top10 = [0] * 10, [0] * 10
    total_scores, efforts, lucks = [0] * 10, [0] * 10, [0] * 10
    
    for _ in range(n):
        effort = round(random.random() * 98, 1)
        luck = round(random.random() * 2, 1)
        total = round(effort + luck, 1)

        if effort > min(effort_top10):
            effort_top10[0] = effort
            effort_top10.sort()

        if luck > min(luck_top10):
            luck_top10[0] = luck
            luck_top10.sort()

        if total > min(total_scores):
            min_idx = total_scores.index(min(total_scores))
            total_scores[min_idx] = total
            efforts[min_idx] = effort
            lucks[min_idx] = luck

        # Uncomment for real-time plotting:
        #clear_output(wait=True)
        #plt.bar(range(10), efforts)
        #plt.bar(range(10), lucks, bottom=efforts)
        #plt.pause(0.01)

    return effort_top10, luck_top10, total_scores, efforts, lucks
```

```python
# Test
n = 100000
effort_top10, luck_top10, total_scores, efforts, lucks = generate_top_scores(n)

# Plot final results
plt.bar(range(10), efforts)
plt.bar(range(10), lucks, bottom=efforts)
plt.show()

# Print results
print('The Winners:', total_scores)
print('The Efforts of the Winners:', efforts)
print('The Lucks of the Winners:', lucks)
print('The 10 Most Efforts:', effort_top10)
```
