##### Success and Luck: Good Fortune and the Myth of Meritocracy

##### 실험
* 0과 98 사이의 난수를 생성하여, 재능/노력에 부여.
* 0과 2 사이의 난수를 생성하여, 행운에 부여.
* 가장 노력한 10명 모두 승자가 되는지 확인해 볼 것.

```python
import random
import matplotlib.pyplot as plt
from IPython.display import clear_output

# test set
n = 100000

# Random number
total_list = [0] * 10

# graph list
effort_list = [0] * 10
luck_list = [0] * 10

# top list
effort_top10 = [0] * 10
luck_top10 = [0] * 10

for _ in range(n):
    # random number
    effort = round(random.random() * 98, 1)
    if effort > min(effort_top10):
        effort_top10[0] = effort
        effort_top10.sort()
    luck = round(random.random() * 2, 1)
    if luck > min(luck_top10):
        luck_top10[0] = luck
        luck_top10.sort()
    total = round(effort + luck, 1)
    if total > min(total_list):
        min_idx = total_list.index(min(total_list))
        total_list[min_idx] = total
        effort_list[min_idx] = effort
        luck_list[min_idx] = luck
    # Realtime Graph
    #clear_output(wait=True)
    #plt.bar(range(len(effort_list)), effort_list) # up
    #plt.bar(range(len(luck_list)), luck_list, bottom=effort_list) # bottom
    #plt.pause(0.01)

plt.bar(range(len(effort_list)), effort_list) # up
plt.bar(range(len(luck_list)), luck_list, bottom=effort_list) # bottom

print('The Winners : ', total_list)
print('the Efforts of the Winners : ', effort_list)
print('the Lucks of the Winners : ', luck_list)

print('The 10 Most efforts : ', effort_top10)
```
